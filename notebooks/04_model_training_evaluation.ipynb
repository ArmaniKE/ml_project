```python
import sys
import os
import matplotlib.pyplot as plt
import seaborn as sns

sys.path.append(os.path.abspath('..'))

from src.preprocessing import clean_and_prepare_data
from src.models import init_models, train_models
from src.evaluation import evaluate_predictions, run_cross_validation

# Load and process data
X_train, X_test, y_train, y_test = clean_and_prepare_data('../data/raw/myntra_products_catalog.csv')

# Cross-Validation
models = init_models()
cv_results = run_cross_validation(models, X_train, y_train, cv=5)

# Train and evaluate
trained_models = train_models(models, X_train, y_train)

names = []
r2_scores = []

for name, model in trained_models.items():
    y_pred = model.predict(X_test)
    metrics = evaluate_predictions(y_test, y_pred)
    names.append(name)
    r2_scores.append(metrics['R2'])

# Visualization of Baseline Results
sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.barplot(x=names, y=r2_scores, hue=names, legend=False, palette="viridis")
plt.title("Baseline Models Comparison (R2 Score)")
plt.ylabel("R2 Score")
plt.ylim(0, 1.0)

for i, val in enumerate(r2_scores):
    plt.text(i, val + 0.02, f"{val:.4f}", ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/raw/myntra_products_catalog.csv')
sns.set_theme(style="whitegrid")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot target distribution after log transformation
log_price = np.log1p(df['Price (INR)'])
sns.histplot(log_price, bins=50, kde=True, ax=axes[0], color='teal')
axes[0].set_title('Log-Transformed Price Distribution')
axes[0].set_xlabel('Log(Price + 1)')

# Plot price vs gender
sns.boxplot(data=df, x='Gender', y=log_price, ax=axes[1], hue='Gender', legend=False, palette='Set2')
axes[1].set_title('Log(Price) Distribution by Gender')
axes[1].set_xlabel('Gender Group')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

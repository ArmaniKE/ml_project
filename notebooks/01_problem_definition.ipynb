# Phase 1: Problem Definition

## Project Objective
The goal of this project is to build a robust machine learning pipeline to predict clothing product prices (`Price (INR)`) using the Myntra e-commerce dataset. 

## Machine Learning Task
This is a **Supervised Learning (Regression)** task. The models map product characteristics to a continuous numerical target variable.

## Selected Methods
To evaluate performance across different learning paradigms, we implement:
* **Baseline/Classic ML:** K-Nearest Neighbors (KNN), Support Vector Regression (SVR), Random Forest Regressor.
* **Deep Learning Extension:** Multi-Layer Perceptron (MLPRegressor).

## Evaluation Metrics
Models are validated using 5-fold Cross-Validation and evaluated on the test set using:
* Mean Absolute Error (MAE)
* Root Mean Squared Error (RMSE)
* Coefficient of Determination ($R^2$)
